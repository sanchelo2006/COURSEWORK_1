```python
#задание 1.1 СОЗДАЕМ отдельно dataframe из таблицы students.csv

students_df = pd.read_csv('students.csv', parse_dates = ['birthday'], encoding = 'utf-8')

#ОЧИЩАЕМ dataframe

students_df['city'].fillna('unknown', inplace = True) #заполняем пустые ячейки в столбце city
                                                      #(чтобы потом не потерять даты)

students_df.dropna(inplace = True) #удаляем строки без дат


students_df['date_birthday'] = pd.to_datetime(students_df['birthday'], errors = 'coerce')#создаем
                                                                                         #вспомогательный столбец


students_df[students_df['date_birthday'].isna()] #выводим проблемные даты на экран и видим,
                                                 #что их все можно удалить


students_df.dropna(inplace = True) #удаляем строки с проблемными датами

students_df.drop('birthday',axis = 1, inplace = True) #удаляем первоначальный столбец с датами



#ВЫЧИСЛЯЕМ заданные метрики возраста всех студентов

now = datetime.datetime.now() #сегодняшняя дата

students_df['age'] = now.year - students_df['date_birthday'].dt.year #создаем столбец с возрастом студентов

students_df = students_df[(students_df['age'] > 6) & (students_df['age'] < 90)] #ограничим возраст промежутком
                                                                                #от 7 до 90 лет (для удаления
                                                                                #возможных выбросов)


print('медианный возраст: ', students_df['age'].median()) #выводим медианный возраст

print()

print('остальные метрики')
students_df['age'].describe() #выводим остальные метрики для всех студентов
```

```python
#задание 1.1 ВЫЧИСЛЯЕМ заданные метрики студентов на каждом курсе

#объединяем даные
analys_df_full = analys_df.merge(students_df, how = 'outer', left_on = 'student_id', right_on = 'id')\
                     .drop(['id_', 'id'], axis = 1)\
                     .rename(columns = {'city':'student_city', 'date_birthday':'student_birthday','age':'student_age'})


students_age_statistics = analys_df_full.groupby(['title', 'student_id']).mean() #группируем по курсу и
                                                                                 #студентам (берем среднее для
                                                                                 #того чтобы скинуть groupby в
                                                                                 #dataframe это не повлияет на значение возраста
                                                                                 #- так как у одного и того же студента возраст один)


students_age_statistics = students_age_statistics.groupby('title') #снова группируем по курсу




print('медианный возраст на разных курсах', students_age_statistics['student_age'].median())

print()

print('остальные метрики')
students_age_statistics['student_age'].describe()
```

```python
#задание 2.1 РАСЧЕТ прироста студентов

#меняем формат даты для удобства использования
analys_df['start_date'] = pd.to_datetime(analys_df["start_date"].dt.strftime('%Y-%m'))
analys_df['finish_date'] = pd.to_datetime(analys_df["finish_date"].dt.strftime('%Y-%m'))

#выбираем только домашние работы
analys_df_grow_students = analys_df[(analys_df['is_homework'] == True)]

#удаляем столбец с finish_date для того чтобы очистить данные от нулевых значений (иначе много данных уйдет)
analys_df_grow_students = analys_df_grow_students.drop('finish_date', axis = 1)
analys_df_grow_students.dropna()

#создаем отдельный датасет только с первыми домашними работами на курсе
analys_df_grow_students_first_hw = analys_df_grow_students.groupby(['title']).agg({'module_number':'min'}).reset_index()
#analys_df_grow_students.head(15)

#объединяем созданный датасет с основным - чтобы получить только первые домашние работы из основного
analys_df_grow_students = analys_df_grow_students_first_hw.merge(analys_df_grow_students)

#подсчитываем количество студентов на каждом курсе в каждом месяце
group_students = analys_df_grow_students.groupby(['title', 'start_date']).count().reset_index()
group_students.head()
```

```python
#задание 2.4 РАСЧЕТ количества прогрессов по домашним работам в каждом месяце

#отбираем только строки с домашними работами

progress_df = analys_df[analys_df['is_homework'] == True]

#отбираем только те столбцы из основного датасета, с которыми будем работать:
#progress_id, status start_date, finish_date

progress_data = progress_df[['progress_id','status', 'start_date', 'finish_date', 'title', 'module_number']]

#форматируем даты (приводим их к виду год - месяц)

progress_data['start_date'] = pd.to_datetime(progress_data["start_date"].dt.strftime('%Y-%m'))
progress_data['finish_date'] = pd.to_datetime(progress_data["finish_date"].dt.strftime('%Y-%m'))

#заполняем пропущенные значения в finish_date по условиям:
#если status = start - ставим 2019-07, если status = fail ставим туже дату что и в start_date
progress_data.loc[progress_data['status'] == 'start', 'finish_date'] = datetime.datetime(2019, 7, 1)
progress_data.loc[progress_data['status'] == 'fail', 'finish_date'] = progress_data['start_date']
progress_data.loc[progress_data['status'] == 'wait', 'finish_date'] = datetime.datetime(2019, 7, 1)

#создаем пустой столбец для заполнения датами (все месяцы из каждого периода процесса)
progress_data['period'] = progress_data.apply(lambda _: '', axis=1)
progress_data['period'] = progress_data['period'].astype('object')

#проходим циклом по нашему датафрейму - вычисляем временные периоды
#переводим datetimeindex в список
#записываем этот список в столбец period
for i in progress_data.index:
    date_range = pd.period_range(progress_data.loc[i,'start_date'],
                               progress_data.loc[i,'finish_date'], freq='M')
    date_range = list(map(lambda x: datetime.datetime.strftime(datetime.datetime.strptime(str(x),'%Y-%m'),'%Y-%m'), date_range))[1:]
  
    #здесь мы еще добавляем начальный месяц периода, так как метод period_range(так же как и date_range)
    #почему то не добавляет эту дату
    date_range.insert(0,progress_data.loc[i,'start_date'])
    progress_data.at[i,'period'] = date_range
    
#раскрываем все списки    
progress_data_period = progress_data.explode('period')

#меняем тип столбца period
progress_data_period['period'] = pd.to_datetime(progress_data_period["period"])

#группируем датасет по курсам и  месяцам и выводим количество процессов
progress_data_final = progress_data_period.groupby(['title','period']).count()['progress_id'].reset_index()

progress_data_final.head()
```

```python
#задание 3.1 РАСЧЕТ характеристик прохождения каждого модуля


#выбираем из основного датасета только выполненные домашние работы
progress_data_modules = analys_df[(analys_df['is_homework'] == True) & (analys_df['status'] == 'done')]
[['title', 'module_title', 'module_number', 'start_date', 'finish_date', 'student_id']]

#удаляем строки с пропущенными значениями
progress_data_modules.dropna()

#группируем данные по курсу, модулу и студентам и создаем новые столбцы с началом выполнения модуля и концом
progress_data_modules = progress_data_modules.groupby(['title',
                                                       'module_title',
                                                       'module_number',
                                                       'student_id']).agg(start_module = ('start_date', 'min'),
                                                                          finish_module = ('finish_date', 'max'))

#создаем столбец с временем прохождения модуля
progress_data_modules['time'] = progress_data_modules['finish_module'] - progress_data_modules['start_module']

#форматируем столбец с временем прохождения модуля (переводим в количество дней)
progress_data_modules['time'] = progress_data_modules.loc[:,'time'].map(lambda x: x.days)

#видим, что там где одно и тоже время в start_date и finish_date -
#показывает минус один день или ноль, поэтому меняем -1 на 1 день
progress_data_modules['time'][(progress_data_modules['time'] == -1) | (progress_data_modules['time'] == 0)] = 1

#создаем новые столбцы с заданными метриками
progress_data_modules_metrics = progress_data_modules.groupby(['title', 'module_title']).agg(min = ('time', 'min'),
                                                           max = ('time', 'max'),
                                                           mean = ('time', 'mean'),
                                                           median = ('time', 'median'))


```

```python
#задание 3.2 ПОСТРОЕНИЕ линейного графика с медианным времененм прохождения каждого модуля для каждого курса


#создаем фигуру
fig = plt.figure(figsize = (10,150))                        

#эта переменная счетчик для курсов
s = 0      

for i in list(progress_data_modules_metrics.index.levels[0]):            # Запуск цикла по каждому значению в списке курсов
    s+=1                                                    
    subplot = fig.add_subplot(15,1,s)                       # Добавление координатной оси
    x_ticks = progress_data_modules_metrics.loc[i].index                 # Присвоение значений по оси Х
    y_ticks = progress_data_modules_metrics.loc[i,'median']         # Присвоение значений по оси У
    x_labels = [i for i in progress_data_modules_metrics.loc[i].index]   # Создание списка подписей для оси X
    subplot.plot(x_ticks, y_ticks)                          # Формирование линейного графика по заданным значениям
    subplot.set_title(i)                                    # становка подписи графика соглсно текущему значению из списка курсо
    plt.xticks(progress_data_modules_metrics.loc[i].index,
               rotation = 'vertical',
               labels = x_labels)        # Установка расположения подписей на оси Х
plt.subplots_adjust(wspace=1, hspace=3)  # Установка расстояния между графиками
plt.show()   
```

```python
#задание 3.3 выявление сезонности медианного времени выполнения домашней работы

#создаем новый датасет (теперь берем все домашние задания для анализа)
progress_data_season = analys_df[(analys_df['is_homework'] == True) & (analys_df['status'] == 'done')]
[['title', 'module_title', 'module_number', 'start_date', 'finish_date', 'student_id']]

#удаляем пропущенные значения
progress_data_season.dropna()

#вычисляем время выполнения каждого домашнего задания
progress_data_season['time'] = progress_data_season['finish_date'] - progress_data_season['start_date']

#форматируем время в дни
progress_data_season['time'] = progress_data_season.loc[:,'time'].map(lambda x: x.days)

#форматируем значения -1 и 0 в единицу
progress_data_season['time'][(progress_data_season['time'] == -1) | (progress_data_season['time'] == 0)] = 1

#форматируем столбец finish_date в месяц
progress_data_season['finish_date'] = progress_data_season.loc[:,'finish_date'].map(lambda x: datetime.datetime.strftime(x,'%m'))

#группируем данные по курсу и месяцу и создаем столбец с медианным временем

progress_data_season = progress_data_season.groupby(['title', 'finish_date']).agg(median_days = ('time', 'median'))

progress_data_season
```

```python
#задание 4.1 ПОСЧИТАТЬ конверсию прехода студентов в следующий модуль

#создаем новый датасет только с домашними заданиями

conversion_df = analys_df[analys_df['is_homework'] == True]

#удаляем пропущенные значения


#берем только работы которые сделаны
conversion_df_finish_task = conversion_df[conversion_df['status'] == 'done']

#находим количество студентов на каждой домашней работе
conversion_df_finish_task = conversion_df_finish_task.groupby(['title','module_number', 'module_title','lesson_number']).\
                                                      agg(students_finish = ('student_id', 'count')).\
                                                      reset_index()

#еще один датасет - находим последнюю домашнюю работу в каждом модуле
students_finish_lesson = conversion_df_finish_task.groupby(['title','module_number', 'module_title']).\
                                                   agg(lesson_number_last = ('lesson_number', 'max')).reset_index()                                                      

#объединяем полученные датасеты
students_finish = students_finish_lesson.merge(conversion_df_finish_task, how = 'left')


#создаем еще один датасет в котором будем искать студентов начавших домашнюю работу
conversion_df_start_task = conversion_df.groupby(['title','module_number', 'module_title','lesson_number']).\
                                         agg(students_start = ('student_id', 'count')).\
                                         reset_index()

#находим первую домашнюю работу в каждом модуле
students_start_lesson = conversion_df_start_task.groupby(['title','module_number', 'module_title']).\
                                                 agg(lesson_number_first = ('lesson_number', 'min')).reset_index()                                                      

#объединяем полученные датасеты
students_start = students_start_lesson.merge(conversion_df_start_task, how = 'left')


#объединение датасетов
conversion = students_start.merge(students_finish, how = 'left')


#сдвигаем столбец с количеством студентов окончивших модуль вниз на одну позицию
conversion.students_finish = conversion.students_finish.shift(1)

#создаем иерархический индекс
conversion = conversion.set_index(['title', 'module_title'])

#удаляем первую строку в каждом курсе
for item in conversion.index.levels[0]:
    conversion = conversion.drop(conversion.loc[item].index[0], level = 1)
    
#считаем конверсию
conversion['conversion'] = conversion['students_start'] / conversion['students_finish']

#удаляем лишние столбцы
conversion.drop(['module_number', 'lesson_number', 'lesson_number_first', 'lesson_number_last'], axis = 1, inplace = True)

conversion.head(20)
```

```python
#задание 4.2 ПОСТРОИТЬ столбчатый график

#создаем фигуру
fig = plt.figure(figsize = (10,150))     

#эта переменная счетчик для курсов
s = 0

for i in list(conversion.index.levels[0]):            # Запуск цикла по каждому значению в списке курсов
    s+=1                                                    
    subplot = fig.add_subplot(15,1,s)                       # Добавление координатной оси
    x_ticks = conversion.loc[i].index                 # Присвоение значений по оси Х
    y_ticks = conversion.loc[i,'conversion']         # Присвоение значений по оси У
    x_labels = [i for i in conversion.loc[i].index]   # Создание списка подписей для оси X
    subplot.bar(x_ticks, y_ticks)                          # Формирование линейного графика по заданным значениям
    subplot.set_title(i)                                    # становка подписи графика соглсно текущему значению из списка курсо
    plt.xticks(conversion.loc[i].index,
               rotation = 'vertical',
               labels = x_labels)   # Установка расположения подписей на оси Х
    #рисование линии единицы (выше - увеличение, ниже уменьшение количества студентов)
    line = subplot.hlines(1,conversion.loc[i].index[0], max(conversion.loc[i].index[-1]), color = 'red') 
plt.subplots_adjust(wspace=1, hspace=3)  # Установка расстояния между графиками
plt.show()   
```

На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

задание 5. Создание метрики успеваемости и на основании этой метрики
создание списков студентов для каждого курса

1. Время прохождения домашнего задания и всего модуля или курса в целом - не подходит для метрики. У разных людей разный распорядок дня, поэтому оыенка будет не показательной.

2. Количество работ со статусом fail - тоже  не подходит, так как некоторые студенты только начали курс, некоторые находятся на  середине и только часть студентов закончили. Можно отсеять студентов, которые не прошли еще полностью курс (у которых не все домашние задания). Для этого нужно взять последний урок последнего модуля с домашним заданием. Определить пограничное число fail и отобрать студентов с большим числом fail.


```python
#по второму пункту выводов по метрике (количество работ со статусом fail)

#выбираем только домашние работы
problem_students = analys_df[analys_df['is_homework'] == True]
                             
#группируем по курсу и модулю. Выбираем только последний урок последнего модуля
problem_students_last_lesson = problem_students.groupby(['title', 'module_number']).agg(lesson_number = ('lesson_number', 'max')).reset_index()

#соединяем датасеты
problem_students = problem_students.merge(problem_students_last_lesson, how = 'left')

#выбираем только статус fail
problem_students = problem_students[problem_students['status'] == 'fail']

#группируем по курсу студентам и считаем количество fail
problem_students_count = problem_students.groupby(['title', 'student_id']).agg(fail_count = ('status', 'count')).reset_index()

#после просмотра полученного датасета - определяем, что количество fail = 2 будет граничным показателем
#отбираем только тех студентов у которых количество fail больше двух
problem_students_count = problem_students_count[problem_students_count['fail_count'] > 2]


#группируем по курсам
problem_students_count = problem_students_count.groupby('title').agg(prob_stud_count = ('student_id', 'count')).reset_index()

#сортируем полученный датасет по убыванию
problem_students_count = problem_students_count.sort_values('prob_stud_count', ascending = False)

problem_students_count.head(15)
                





```
